## Método de Euler



```python
def euler( func, h, tini, tfin, x0, params):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):
    
      x.append (xi)
      t.append (ti)
  
      xi = xi + h*func(ti,xi,params)
      ti = ti + h
    return t, x    
```

1.Lineal

```python
def lineal(t, x, a):
    x_dot = a*x
    return x_dot
```

```python
a = 0.5
h = 0.1
tini = 0
tfin = 10
x0 = 0.8


t1,x1 = euler( lineal, h, tini, tfin, x0, a )

plt.figure()
# gráfica de analítica:
#plt.plot(  t, x, 'black', label = 'Analytical', linewidth = 2)
# gráfica de euler:
plt.scatter( t1, x1, c = 'crimson', label = 'Euler', s = 100)
plt.xlim([0, tfin])
#plt.ylim([-3.14, 6.28])
plt.xlabel(r'$t$')
plt.grid()
plt.show()

for i in range(-9,9,1):
    x0=i
    t1,x1 = euler( lineal, h, tini, tfin, x0, a )
    plt.plot( t1, x1, c = 'crimson')

plt.xlim([0, tfin])
#plt.ylim([-3.14, 6.28])
plt.xlabel(r'$t$')
plt.grid()
plt.show()
```

# 3. Circuito RC (Lineal)

```python

def Circuito_RC(t, q, a):
    q_dot = -q/(a)
    return q_dot
```

# 4. Ecuación logística

```python

def Ecuacion_log(t, p, a):
    p_dot = 0.02*p*(1-p/a)
    return p_dot
```

```python

k=100
a=k

h = 10
tini = 0
tfin = 1000
x0 = 0.1


t1,x1 = euler( Ecuacion_log, h, tini, tfin, x0, a )

plt.figure()
# gráfica de analítica:
#plt.plot(  t, x, 'black', label = 'Analytical', linewidth = 2)
# gráfica de euler:
plt.scatter( t1, x1, c = 'crimson', label = 'Euler', s = 100)
plt.xlim([0, tfin])
#plt.ylim([-3.14, 6.28])
plt.xlabel(r'$t$')
plt.grid()
plt.show()

for i in range(-100,100,10):
    x0=i
    t1,x1 = euler( Ecuacion_log, h, tini, tfin, x0, a )
    plt.plot( t1, x1, c = 'crimson')

plt.xlim([0, 500])
plt.ylim([0, 110])
plt.xlabel(r'$t$')
plt.grid()
plt.show()
```
